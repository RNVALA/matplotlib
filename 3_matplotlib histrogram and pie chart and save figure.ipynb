```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blood_sugar=[113,85,90,150,88,115,135,80,77,82,129]
plt.hist(blood_sugar,bins=[80,100,125,200],rwidth=0.95,color='g')#histtype='step'
#in hisrogram there is concept called bin and bucket 
```

```python
blood_sugar_women=[67,98,89,120,133,150,84,69,89,79,120,112,100]
plt.xlabel("sugar range")
plt.ylabel('total no of patient')
plt.title('blood sugar analytics')
plt.hist([blood_sugar,blood_sugar_women],bins=[80,100,125,150],rwidth=0.95,color=['g','r'],label=['men','women'])
#by orientation we can put histrogram horizontal syntax is orientation='horizontal'
plt.legend()
```

```python
blood_sugar_women=[67,98,89,120,133,150,84,69,89,79,120,112,100]
plt.xlabel("total no of patient")
plt.ylabel('sugar range')
plt.title('blood sugar analytics')
plt.hist([blood_sugar,blood_sugar_women],bins=[80,100,125,150],rwidth=0.95,color=['g','r'],label=['men','women'],orientation='horizontal')
#by orientation we can put histrogram horizontal syntax is orientation='horizontal'
plt.legend()
```

### pie chart

```python
exp_vals=[1400,600,300,410,250]
exp_labels=['Home Rent','Food','phone/internet Bill','car','other utilities']
```

```python
plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels,radius=2.5,autopct='%0.1f%%',shadow=True,explode=[0,0.2,0.5,0,0],startangle=90)
plt.show()
#here autocpt will print the percentage of pie chart , shadow will give shadow , explode will take about portion pie out and startangle is from which angle you want start of first values
```

## saving chart

```python
plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels,radius=2.5,autopct='%0.1f%%',shadow=True,explode=[0,0.2,0.5,0,0],startangle=90)
# plt.show()

plt.savefig('pie.png',bbox_inches='tight',pad_inches=2,transparent=True)
#here b_box_inches means it will store entire chart with labels
#pad_inches-if i want to provide a padding 
#transparent= True means the only fig will show backgroup will be waived off
```
