## What is matplotlib
- Matplotlib is a python 2d plotting library it is primarily for data visuvalization


### - Topics covered
- plot graph
- fromat string in plot function
- Axies,labels,legend,grid

```python
import matplotlib.pyplot as plt#pyplot is main object  you got from matplotlob
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,52,57,32]
```

```python
plt.title('temperature graph')
plt.xlabel('temperature',fontsize=20,color='red')
plt.ylabel('days',fontsize=20,color='blue')
plt.plot(x,y,color='green',linestyle='dashed',marker='o')
#it have optional parameter that is color='green',linestyle='dashed',marker='o',linewidth=5
```

### format string in plot function

```python
plt.plot(x,y,'--r*')
plt.plot(x,y,'rD')#color=red, marker=diamind
```

```python
plt.plot(x,y,color='#888888',marker='D',linestyle='',markersize=20)
```

```python
plt.plot(x,y,color='green',alpha=0.2)#alpha ranges from 0 to 1,0 means not visible 
```

## Axies,labels,legend,grid

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
#in this plot we plot a labels additional to previous plot and we are using legend for showing the labels

#if you want to specify that i want labels on right corner than syntax is plt.legend(loc='upper right') and bydefault it fitted in best 
#for location we  have best, upper right, upper left,lower left,lower right,right,center left,center right,lower center,upper center,center

plt.xlabel('Days')
plt.ylabel('temperature')
plt.title('weather of different days')
plt.plot(days,max_t,label='max_temperature')
plt.plot(days,min_t,label='min_temperature')
plt.plot(days,avg_t,label='average temperature')

plt.legend(loc="upper right")
#we can also past that shadow=True,fontsize='large'
plt.grid()#it will show the grid so we can visuvalize the point better
```
